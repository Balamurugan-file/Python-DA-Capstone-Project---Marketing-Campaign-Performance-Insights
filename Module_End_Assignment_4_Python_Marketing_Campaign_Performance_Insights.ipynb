###3) Exploratory Data Analysis (EDA) and Visualization

####Campaign Performance:

```python
# Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI.

plt.figure(figsize=(15,6))

sns.scatterplot(data=df,x='Acquisition_Cost',y='ROI')

plt.title('Relationship between Acquisition_Cost and ROI')
plt.xlabel('Acquisition Cost')
plt.ylabel('ROI')
plt.tight_layout()
plt.show()
```

```python
# Create a bar chart to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type.

plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Channel_Used',y='Conversion_Rate',hue='Campaign_Type')

plt.title('Average Conversion Rate by Channel Used and Campaign Type')
plt.xlabel('Channel Used')
plt.ylabel('Average Conversion Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
'''
Visualize the distribution of Engagement_Score across different Campaign_Type
using a box plot.
'''
sns.boxplot(data=df,x='Campaign_Type',y='Engagement_Score',hue='Campaign_Type')

plt.title('Distribution of Engagement_Score across Campaign_Type')
plt.xlabel('Campaign Type')
plt.ylabel('Engagement Score')
plt.xticks(rotation=45)
plt.show()
```

```python
'''
Analyze the average ROI by Company using a bar chart to compare the profitability
of campaigns conducted by different companies.
'''

plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Company',y='ROI',hue='Company')

plt.title('Average ROI by Company')
plt.xlabel('Company')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
'''
Examine the correlation between Engagement_Score and Conversion_Rate
using a heatmap.
'''

coorelation_matrix = df[['Engagement_Score','Conversion_Rate']].corr()
plt.figure(figsize=(10,5))

sns.heatmap(coorelation_matrix,annot=True)

plt.title('Correlation b/w Engagement_Score and Conversion_Rate')
plt.xticks(rotation=45)
plt.yticks(rotation=45)

plt.tight_layout()
plt.show()
```

###Customer Segmentation:

```python
# Create a count plot to visualize the distribution of Target_Audience.
plt.figure(figsize=(10,5))

ax = sns.countplot(data=df,x='Target_Audience')
ax.bar_label(ax.containers[0])

plt.title('Distribution of Target_Audience')
plt.xlabel('Target Audience')
plt.ylabel('Count')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
'''
Identify which Customer_Segment has the highest Conversion_Rate for each
Language using a bar chart.
'''

grouped = df.groupby(['Language', 'Customer_Segment'])['Conversion_Rate'].mean().reset_index()


plt.figure(figsize=(10,8))
sns.barplot(data=grouped,x='Language',y='Conversion_Rate',hue='Customer_Segment')

plt.title('Total fare for each pickup_borough')
plt.xlabel('pickup_borough')
plt.ylabel('Total fare')

plt.tight_layout()
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
'''
Visualize the distribution of Acquisition_Cost across each Customer_Segment,
categorized by Channel_Used, using a box plot.
'''

plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='Customer_Segment', y='Acquisition_Cost', hue='Channel_Used')

plt.title('Distribution of Acquisition Cost across Customer Segment by Channel Used')
plt.xlabel('Customer Segment')
plt.ylabel('Acquisition Cost')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
'''
Analyze average Conversion_Rate by Language using a bar chart to compare
the effectiveness of campaigns conducted in different languages.
'''

plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Language',y='Conversion_Rate',hue='Language')

plt.title('Average Conversion Rate by Language')
plt.xlabel('Language')
plt.ylabel('Average Conversion Rate')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###Channel Effectivenes:

```python
'''
Compare the Engagement_Score for different Channels_Used, segmented by
Campaign_Type, using a bar chart.
'''
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Channel_Used',y='Engagement_Score',hue='Campaign_Type')

plt.title('Engagement_Score for different Channels used by Campaign Type')
plt.xlabel('Channel Used')
plt.ylabel('Engagement_Score')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Show the distribution of total ROI across different Channels_Used using a pie chart.

ROI_count = df.groupby('Channel_Used')['ROI'].sum()
ROI_count.plot(kind='pie',autopct='%1.2f%%',figsize=(5,5),startangle=90,shadow=True)

plt.title('Distribution of total ROI across different Channels Used')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
'''
Plot a scatter plot to show the relationship between Clicks and Impressions
for each Channel_Used.
'''
plt.figure(figsize=(10,8))
sns.scatterplot(data=df,x='Clicks',y='Impressions',hue='Channel_Used',alpha=0.8)

plt.title('Relationship Between Clicks and Impressions for each Channel_Used')
plt.xlabel('Clicks')
plt.ylabel('Impressions')

plt.tight_layout()
plt.show()
```

###Time-Based Analysis:

```python
# Plot the distribution of Duration using a histogram.

plt.figure(figsize=(10,6))
plt.hist(df['Duration'],bins=20,edgecolor='k',color='g')

plt.title('Distribution of Duration')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
'''
Analyze how the overall Conversion_Rate has changed over Date for each
Company using a line chart.
'''
grouped = df.groupby(['Date', 'Company'])['Conversion_Rate'].mean().reset_index()

plt.figure(figsize=(20,6))
sns.lineplot(data=grouped, x='Date', y='Conversion_Rate', hue='Company', marker='o')

plt.title('Conversion Rate Over Time for Each Company')
plt.xlabel('Date')
plt.ylabel('Conversion Rate')
plt.xticks(rotation=90)
plt.tight_layout()


plt.show()
```

```python
# Examine the trend of Engagement_Score over Date with a line chart.
plt.figure(figsize=(20,6))
grouped = df.groupby('Date')['Engagement_Score'].mean().reset_index()

sns.lineplot(data=grouped, x='Date', y='Engagement_Score',marker='o')

plt.title('Trend of Engagement Score Over Time')
plt.xlabel('Date')
plt.ylabel('Average Engagement Score')

plt.xticks(rotation=90)
plt.tight_layout()


plt.show()
```

###Geographic Analysis:

```python
# Determine which location has the highest Acquisition_Cost using a bar chart.

grouped = df.groupby('Location')['Acquisition_Cost'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=grouped, x='Location', y='Acquisition_Cost',hue='Location')

plt.title('Highest Acquisition_Cost by location')
plt.xlabel('Location')
plt.ylabel('Acquisition_Cost')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
'''
Visualize the Conversion_Rate by different Location, categorized by
Target_Audience, using a bar chart.
'''

plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Location', y='Conversion_Rate',hue='Target_Audience')

plt.title('Conversion Rate by Location Categorized by Target Audience')
plt.xlabel('Location')
plt.ylabel('Conversion_Rate')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Illustrate the proportion of ROI by Location using a pie chart.

# Sum of ROI by Location
roi_by_location = df.groupby('Location')['ROI'].sum()

plt.figure(figsize=(6,6))
plt.pie(roi_by_location,labels=roi_by_location.index,autopct='%1.1f%%',startangle=90,shadow=True)


plt.title('Proportion of ROI by Location')

plt.tight_layout()
plt.show()
```
